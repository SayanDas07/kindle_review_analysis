```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/drive/MyDrive/Colab Notebooks/kindle_review_NLP/all_kindle_review.csv")
```

```python
df = df[['reviewText','rating']]
df.head()
```

```python
df['rating'].value_counts()
```

```python
df['rating']=df['rating'].apply(lambda x:0 if x<3 else 1)
```

```python
df.head()
```

```python
df['rating'].value_counts()
```

```python
df['reviewText']=df['reviewText'].str.lower()
```

```python
df.head()
```

```python
import re
import nltk
from nltk.corpus import stopwords
from bs4 import BeautifulSoup
nltk.download('stopwords')
```

```python
for i in range(0,len(df["reviewText"])):
    text = re.sub('[^a-z A-Z 0-9-]+', '', df['reviewText'][i])

    # Remove stopwords
    text = " ".join([word for word in text.split() if word not in stopwords.words('english')])

    # Remove URLs
    text = re.sub(r'(http|https|ftp|ssh)://([\w_-]+(?:(?:\.[\w_-]+)+))([\w.,@?^=%&:/~+#-]*[\w@?^=%&/~+#-])?', '', text)

    # Remove HTML tags
    text = BeautifulSoup(text, 'lxml').get_text()

    # Remove any additional spaces
    text = " ".join(text.split())

    # Update the DataFrame
    df.at[i, 'reviewText'] = text
```

```python
df.head()
```

```python
from nltk.stem import WordNetLemmatizer
```

```python
nltk.download('wordnet')
```

```python
lemmatizer=WordNetLemmatizer()
```

```python
def lemmatize_words(text):
    return " ".join([lemmatizer.lemmatize(word) for word in text.split()])
```

```python
df['reviewText']=df['reviewText'].apply(lambda x:lemmatize_words(x))
```

```python
from sklearn.model_selection import train_test_split

X_train,X_test,y_train,y_test=train_test_split(df['reviewText'],df['rating'],
                                              test_size=0.20)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
bow=CountVectorizer()
X_train_bow=bow.fit_transform(X_train).toarray()
X_test_bow=bow.transform(X_test).toarray()
```

```python
from sklearn.naive_bayes import GaussianNB
nb_model_bow=GaussianNB().fit(X_train_bow,y_train)
```

```python
from sklearn.metrics import confusion_matrix, accuracy_score
y_pred = nb_model_bow.predict(X_test_bow)
cm = confusion_matrix(y_test, y_pred)
print(cm)
accuracy_score(y_test, y_pred)
```
